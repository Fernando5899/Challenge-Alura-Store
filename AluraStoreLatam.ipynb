# 4. Productos más y menos vendidos

## En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

### **Tienda 1**

```python
# Guardamos los datos de la primera tienda en una variable
ventas_1 = tienda_1['Producto']
ventas_1.value_counts()
```

```python
# Guardamos el conteo en una variable
conteo_ventas_1 = ventas_1.value_counts()
```

```python
# Obtenemos cual es el número maximo de ventas
maximo_ventas_1 = conteo_ventas_1.max()

# Filtramos para obtener solo las filas con el maximo de ventas
productos_mas_vendidos_1 = conteo_ventas_1[conteo_ventas_1 == maximo_ventas_1].index.tolist()

# Imprimimos el resultado
print(f"Los productos más vendidos de la tienda 1 son: {productos_mas_vendidos_1}")
```

```python
# Obtenemos cual es el numero de ventas menor de la tienda 1
minimo_ventas_1 = conteo_ventas_1.min()

# Filtramos para obtener solo las filas con el numero de ventas menor
productos_menos_vendidos_1 = conteo_ventas_1[conteo_ventas_1 == minimo_ventas_1].index.tolist()

# Imprimimos los productos menos vendidos
print("Productos menos vendidos en la tienda 1:", productos_menos_vendidos_1)
```

### **Tienda 2**

```python
ventas_2 = tienda_2['Producto']
ventas_2.value_counts()
```

```python
# Buscando los productos más vendidos en la tienda 2
conteo_ventas_2 = ventas_2.value_counts()

# Buscando cuál es el número máximo de ventas
maximo_ventas_2 = conteo_ventas_2.max()

# Filtrando los productos más vendidos de la tienda 2
productos_mas_vendidos_2 = conteo_ventas_2[conteo_ventas_2 == maximo_ventas_2].index.tolist()

# Imprimiendo los productos más vendidos de la tienda 2
print(f"Los productos más vendidos en la tienda 2 son: {productos_mas_vendidos_2}")
```

```python
# Buscamos el minimo de ventas
minimo_ventas_2 = conteo_ventas_2.min()

# Filtrando los productos menos vendidos de la tienda 2
productos_menos_vendidos_2 = conteo_ventas_2[conteo_ventas_2 == minimo_ventas_2].index.tolist()

# Imprimiendo los productos menos vendidos de la tienda 2
print(f"Los productos menos vendidos de la tienda 2 son: {productos_menos_vendidos_2}")
```

### **Tienda 3**

```python
ventas_3 = tienda_3['Producto']
ventas_3.value_counts()
```

```python
# Vemos el total de ventas de cada producto dentro de la tienda 3
conteo_ventas_3 = ventas_3.value_counts()

# Viendo cual es el máximo de ventas en la tienda 3
maximo_ventas_3 = conteo_ventas_3.max()

# Viendo el minimo de ventas en la tienda 3
minimo_ventas_3 = conteo_ventas_3.min()

# Filtrando los productos mas vendidos
productos_mas_vendidos_3 = conteo_ventas_3[conteo_ventas_3 == maximo_ventas_3].index.tolist()

# Filtrando los productos menos vendidos
productos_menos_vendidos_3 = conteo_ventas_3[conteo_ventas_3 == minimo_ventas_3].index.tolist()

# Imprimiendo los resultados de nuestros filtros
print(f"Los productos más vendidos de la tienda 3 son: {productos_mas_vendidos_3}")
print(f"Los productos menos vendidos de la tienda 3 son: {productos_menos_vendidos_3}")

```

### **Tienda 4**

```python
ventas_4 = tienda_4['Producto']
ventas_4.value_counts()
```

```python
# Guardamos en una variable el conteo total de ventas de la tienda 4
conteo_ventas_4 = ventas_4.value_counts()

# Viendo cual es el número mayor de ventas de la tienda 4
maximo_ventas_4 = conteo_ventas_4.max()

# Viendo cual es el menor número de ventas de la tienda 4
minimo_ventas_4 = conteo_ventas_4.min()

# Filtrando los datos para ver cuales son los productos mas vendidos de la tienda 4
productos_mas_vendidos_4 = conteo_ventas_4[conteo_ventas_4 == maximo_ventas_4].index.tolist()

# Filtrando los datos para ver cuales son los productos menos vendidos de la tienda 4a
productos_menos_vendidos_4 = conteo_ventas_4[conteo_ventas_4 == minimo_ventas_4].index.tolist()

# Imprimiendo los productos mas vendidos de la tienda 4
print(f"Productos mas vendidos de la tienda 4: {productos_mas_vendidos_4}")

# Imprimiendo los productos menos vendidos de la tienda 4
print(f"Productos menos vendidos de la tienda 4: {productos_menos_vendidos_4}")
```
