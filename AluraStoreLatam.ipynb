### Importación de datos



```python
import pandas as pd

# Ponemos en una variable la url de cada csv
url_1 = 'base-de-datos-challenge1-latam/tienda_1.csv'
url_2 = 'base-de-datos-challenge1-latam/tienda_2.csv'
url_3 = 'base-de-datos-challenge1-latam/tienda_3.csv'
url_4 = 'base-de-datos-challenge1-latam/tienda_4.csv'

# Poniendo en una variable los datos para que pandas lo pueda leer
tienda_1 = pd.read_csv(url_1)
tienda_2 = pd.read_csv(url_2)
tienda_3 = pd.read_csv(url_3)
tienda_4 = pd.read_csv(url_4)

# Leyendo los datos de la primera tienda
tienda_1.head()
```

# 1. Análisis de facturación

# 2. Ventas por categoría

## En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

### **Tienda 1**

```python
cantidad_productos_vendidos_1 = tienda_1['Categoría del Producto']
cantidad_productos_vendidos_1.value_counts()
```

```python
# Mostrando cuál es la categoria más popular
tienda_1_categoria_mas_popular = cantidad_productos_vendidos_1.value_counts().idxmax()
print(f"La categoría más popular de la Tienda 1 es: {tienda_1_categoria_mas_popular}")
```

# 3. Calificación promedio de la tienda


## En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

### **Tienda 1**

```python
calificacion_1 = tienda_1['Calificación']
calificacion_1
```

```python
# Calculando las calificaciones promedio de la tienda 1
calificacion_promedio_1 = calificacion_1.mean()
print(f"La calificación promedio de la Tienda 1 es: {calificacion_promedio_1.__round__(3)}")
```
