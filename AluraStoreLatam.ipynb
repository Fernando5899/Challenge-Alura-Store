# 2. Ventas por categoría

## En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

### **Tienda 1**

```python
cantidad_productos_vendidos_1 = tienda_1['Categoría del Producto']
cantidad_productos_vendidos_1.value_counts()
```

```python
# Mostrando cuál es la categoria más popular
tienda_1_categoria_mas_popular = cantidad_productos_vendidos_1.value_counts().idxmax()
print(f"La categoría más popular de la Tienda 1 es: {tienda_1_categoria_mas_popular}")
```
