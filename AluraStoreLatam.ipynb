# 4. Productos más y menos vendidos

## En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

### **Tienda 1**

```python
# Guardamos los datos de la primera tienda en una variable
ventas_1 = tienda_1['Producto']
ventas_1.value_counts()
```

```python
# Guardamos el conteo en una variable
conteo_ventas_1 = ventas_1.value_counts()
```

```python
# Obtenemos cual es el número maximo de ventas
maximo_ventas_1 = conteo_ventas_1.max()

# Filtramos para obtener solo las filas con el maximo de ventas
productos_mas_vendidos_1 = conteo_ventas_1[conteo_ventas_1 == maximo_ventas_1].index.tolist()

# Imprimimos el resultado
print(f"Los productos más vendidos de la tienda 1 son: {productos_mas_vendidos_1}")
```

```python
# Obtenemos cual es el numero de ventas menor de la tienda 1
minimo_ventas_1 = conteo_ventas_1.min()

# Filtramos para obtener solo las filas con el numero de ventas menor
productos_menos_vendidos_1 = conteo_ventas_1[conteo_ventas_1 == minimo_ventas_1].index.tolist()

# Imprimimos los productos menos vendidos
print("Productos menos vendidos en la tienda 1:", productos_menos_vendidos_1)
```

```python
# Grafica para ver los ingresos por tienda

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Definimos los datos 
nombre_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas_por_tienda = [total_ingreso_tienda_1, total_ingreso_tienda_2, total_ingreso_tienda_3, total_ingreso_tienda_4]
colores_barras = ['skyblue', 'lightgreen', 'orange', 'pink']

# Configuración del tamaño de la gráfica
plt.figure(figsize=(12, 7))

# Creando el gráfico
grafico = plt.bar(nombre_tiendas, ventas_por_tienda, color=colores_barras)

# Personalización del gráfico
plt.title('Comparación de Ingresos por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=14)
plt.ylabel('Ingresos (en Millones de $)', fontsize=14)

# Función para formatear: divide por 1 millón y agrega 'M'
def formato_millones(x, pos):
    return f'${x / 1_000_000:,.0f} M'

# Aplicamos el formato al eje Y
formato = FuncFormatter(formato_millones)
plt.gca().yaxis.set_major_formatter(formato)

# Añadimos el valor exacto encima de cada barra
for barra in grafico:
    altura = barra.get_height()
    # Aquí dejamos el valor completo para tener la cantidad exacta
    plt.text(barra.get_x() + barra.get_width() / 2, altura, 
             f'${altura:,.0f}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Visualizando el gráfico
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(0, max(ventas_por_tienda) * 1.15)
plt.show()
```

```python
# Acomodando datos para las siguientes graficas :)
import seaborn as sns

if 'Tienda' not in tienda_1.columns:
    tienda_1['Tienda'] = 'Tienda 1'
    tienda_2['Tienda'] = 'Tienda 2'
    tienda_3['Tienda'] = 'Tienda 3'
    tienda_4['Tienda'] = 'Tienda 4'

    df_final = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4])
    
    # Conversión de las fechas
    df_final['Fecha de Compra'] = pd.to_datetime(df_final['Fecha de Compra'], dayfirst=True)
    df_final['Mes_Año'] = df_final['Fecha de Compra'].dt.to_period('M')
```

```python
plt.figure(figsize=(12, 7))

# Calculamos
ticket_promedio = df_final.groupby('Tienda')['Precio'].mean().reset_index()

# Graficamos
ax = sns.barplot(data=ticket_promedio, x='Tienda', y='Precio', hue='Tienda', legend=False, palette='viridis', edgecolor=".2")

# Títulos y Estilo
plt.title('Ticket Promedio de Venta por Tienda', fontsize=16, fontweight='bold')
plt.ylabel('Valor Promedio ($)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)

# Etiquetas de valor automático
for i in ax.containers:
    ax.bar_label(i, fmt='$%.0f', padding=3, fontsize=11, fontweight='bold')

sns.despine(left=True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

df_final['Ratio_Envio'] = df_final['Costo de envío'] / df_final['Precio']
ratio_envio = df_final.groupby('Tienda')['Ratio_Envio'].mean().reset_index()
ratio_envio['Ratio_Envio'] = ratio_envio['Ratio_Envio'] * 100 # A porcentaje

ax = sns.barplot(data=ratio_envio, y='Tienda', x='Ratio_Envio', hue='Tienda', legend=False,  palette='Reds_r', edgecolor=".2")

plt.title('Impacto del Costo de Envío (% sobre el Precio)', fontsize=16, fontweight='bold')
plt.xlabel('Porcentaje (%)', fontsize=12)
plt.ylabel('')

# Etiquetas
for i in ax.containers:
    ax.bar_label(i, fmt='%.2f%%', padding=5, fontsize=11, fontweight='bold')

plt.xlim(0, max(ratio_envio['Ratio_Envio']) * 1.2)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))

sns.boxplot(data=df_final, x='Tienda', y='Calificación', hue='Tienda', legend=False,  palette='coolwarm', linewidth=2, fliersize=5)

plt.title('Calificación de la Satisfacción del Cliente', fontsize=16, fontweight='bold')
plt.ylabel('Calificación (1-5 Estrellas)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
df_final['Fecha de Compra'] = pd.to_datetime(df_final['Fecha de Compra'], dayfirst=True)

# Creamos la columna del periodo (Mes-Año)
df_final['Mes_Año'] = df_final['Fecha de Compra'].dt.to_period('M')

# 2. Ahora sí graficamos
plt.figure(figsize=(14, 8))

ventas_mensuales = df_final.groupby(['Mes_Año', 'Tienda'])['Precio'].sum().reset_index()
ventas_mensuales['Mes_Año_Str'] = ventas_mensuales['Mes_Año'].astype(str)
ventas_mensuales = ventas_mensuales.sort_values(by='Mes_Año')

sns.lineplot(data=ventas_mensuales, x='Mes_Año_Str', y='Precio', hue='Tienda', 
             style='Tienda', markers=True, dashes=False, linewidth=3, palette='tab10', markersize=9)

plt.title('Tendencia de Ingresos Mensuales (Evolución)', fontsize=16, fontweight='bold')
plt.ylabel('Ingresos Totales ($)', fontsize=12)
plt.xlabel('Periodo', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```
